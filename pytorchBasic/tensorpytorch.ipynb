```python
a = [1, 4, 5, 3, 2]
x = torch.tensor(a, dtype=torch.int)
x
```

```python
# cong 2 phan tu
x = torch.ones(5, 3, dtype=torch.long)
y = torch.randn_like(x, dtype=torch.float)
z = x + y
z
```
